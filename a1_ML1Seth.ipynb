## Types of Machine Learning Systems


Machine Learning (ML) systems can be grouped by how they learn, adapt, and generalize:

---

### 1. Based on Supervision

- **Supervised Learning** uses labeled data to predict outcomes (classification or regression).  
  **Examples:** Spam detection, housing price prediction.  
  **Algorithms:** Logistic Regression, Decision Trees, Neural Networks.

- **Unsupervised Learning** finds patterns in unlabeled data (e.g., clustering, anomaly detection, association rules).  
  **Examples:** Customer segmentation, fraud detection.  
  **Algorithms:** k-Means, PCA, t-SNE, Apriori.

- **Semisupervised Learning** combines a small amount of labeled data with large unlabeled sets.  
  **Example:** Photo tagging apps that use face clustering plus a few labels.

- **Reinforcement Learning** trains agents through interactions with environments using reward signals.  
  **Examples:** AlphaGo, autonomous robots.

---

### 2. Based on Learning Method

- **Batch Learning** trains once on the full dataset; resource-intensive but stable.

- **Online Learning** updates continuously as new data arrives; useful for real-time systems and large-scale applications.  
  **Note:** Learning rate tuning is critical—high rates adapt fast but forget easily; low rates adapt slowly but maintain stability.

---

### 3. Based on Generalization

- **Instance-Based Learning** stores training data and compares new inputs using similarity (e.g., k-NN).

- **Model-Based Learning** builds a model that generalizes from training data.  
  **Process:** Choose model → define cost function → train by minimizing error (e.g., Linear Regression for GDP vs. life satisfaction).

---

### Other Key Concepts

- **Dimensionality Reduction** (e.g., PCA, LLE) compresses data while preserving structure, aiding performance and visualization.

- **Anomaly Detection** flags rare, unusual data points—important in fraud detection, manufacturing, or data cleaning.

- **Association Rule Learning** uncovers relationships in data (e.g., customers who buy BBQ sauce also buy steak).

---

> ML categories often overlap for example, a deep learning spam filter may be supervised, online, and model-based.


```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing

# Load built-in dataset
data = fetch_california_housing(as_frame=True)
df = data.frame[["MedInc", "MedHouseVal"]].copy()
df.columns = ["GDP per capita", "Life satisfaction"]

# Prepare data
X = df[["GDP per capita"]].values
y = df[["Life satisfaction"]].values

# Visualize
df.plot(kind='scatter', x="GDP per capita", y="Life satisfaction", alpha=0.3)
plt.title("GDP vs Life Satisfaction")
plt.grid(True)
plt.show()

# Train linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict life satisfaction for a new GDP value
X_new = [[8.0]]  # Example GDP per capita
prediction = model.predict(X_new)
print("Predicted Life Satisfaction for GDP per capita 8.0:", prediction[0][0])
```

## Main Challenges of Machine Learning


Machine Learning (ML) systems can fail due to two major factors: bad data and bad algorithms.

---

### 🔹 Bad Data Issues

#### **Insufficient Training Data**
ML needs far more examples than humans to learn. Even simple tasks may require thousands of examples, while tasks like image or speech recognition may require millions.

#### **Nonrepresentative Data**
If the training data doesn't reflect the real-world distribution (e.g., missing out on certain groups), the model won't generalize well.  
**Example:** GDP vs. life satisfaction—missing a few countries changes the regression line drastically.

> **Sampling bias** occurs when your data collection method is flawed.  
> **Historical example:** The 1936 US election poll by *Literary Digest* predicted the wrong winner due to sampling bias.

#### **Poor-Quality Data**
Errors, noise, and outliers can mislead the model.  
Data cleaning is often the most time-consuming but essential task in ML.

#### **Irrelevant Features**
The system won’t learn well if data includes too many useless features.  
**Feature engineering** helps by:
- Selecting useful features
- Creating new features
- Using dimensionality reduction

---

### 🔹 Bad Algorithm Issues

#### **Overfitting**
The model learns the training data too well—including noise and outliers—leading to poor generalization.  
**Example:** A complex model might believe countries with a "W" in their name are happier, based purely on coincidence.

**Solutions:**
- Use simpler models
- Get more (and cleaner) data
- Apply regularization (limit model flexibility via hyperparameters)

#### **Underfitting**
The model is too simple to learn the underlying patterns.

**Fixes:**
- Use a more complex model
- Improve features
- Reduce regularization

## 16. What is a test set and why would you want to use it?

A test set is a subset of the dataset not used during training. It’s used to estimate how well the model will generalize to new, unseen data.

---

## 17. What is the purpose of a validation set?

To evaluate and fine-tune model hyperparameters without touching the test set, ensuring the test set remains a fair, unbiased measure of generalization.

---

## 18. What can go wrong if you tune hyperparameters using the test set?

The model may become biased to the test set, leading to overly optimistic performance estimates that don’t generalize to real-world data.

---

## 19. What is cross-validation and why would you prefer it to a validation set?

Cross-validation is a technique where the training data is split into multiple folds.  
The model is trained and validated multiple times, each time using a different fold as the validation set.  
It gives a more reliable estimate of model performance, especially when data is limited.
